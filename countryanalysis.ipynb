```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import math
from __future__ import print_function
from IPython.core.display import HTML
# This just lets the output of the following code samples
#  display inline on this page.%pylab inline
%pylab inline
from pylab import *

# Set default figure size for this notebook
pylab.rcParams['figure.figsize'] = (8.0, 6.4)

def loadTable():
    path = 'C:\Stuffs\StonyBrook\Courses\DataScience\country_data2.csv'
    datatable = pd.read_table(path, sep=',', header=0)
    for k, v in datatable.iloc[:,13:21].convert_objects(convert_numeric=True).iteritems():
        datatable[k] = v
    datatable = datatable.fillna(datatable.mean())
    return datatable

def loadTable1():
    path = 'C:\Stuffs\StonyBrook\Courses\DataScience\country_data2.csv'
    datatable = pd.read_table(path, sep=',', header=0)
    for k, v in datatable.iloc[:,13:21].convert_objects(convert_numeric=True).iteritems():
        datatable[k] = v
    return datatable

#Reading column names from datatable
#colNames = [col for col in datatable]

#Reading a cell from the datatable by specifying the column and the row no
#datatable[colNames[0]][1]

#Value count - Stores no. of occurances of a given key in HashMap
#populationCount = datatable['Government Type'].value_counts()

#Plot a simple graph
#print populationCount[:10]
#populationCount[:10].plot(kind='barh',rot=0)

# Computing correlation factor between all columns
candidatesList = ['Population','Life Exp','GDP (PPP) in US $','Land Boundaries (km)',
                        'Coastline (km)','Literacy Rate (%)','Health Exp/GDP','Military Exp/GDP',
                        'Renewable Water Resources (cu km)','Net Migration Rate/1000 Population',
                        'Labor Force','Internet Users (circa 2009)','Commercial Prime Lending Rate']




def computeDataTableCorr(datatable, candidatesList):
    corrCandidates =  datatable[candidatesList]
    return corrCandidates.corr()

# Plotting correlation heat graph
def displayCorrHeatGraph(cTable, title):
    #_ = pd.scatter_matrix(corrTable, diagonal='kde', figsize=(10, 10))
    plt.imshow(cTable, cmap='hot', interpolation='none')
    plt.colorbar()
    plt.xticks(range(len(cTable)), cTable.columns, rotation=90)
    plt.yticks(range(len(cTable)), cTable.columns)
    plt.title(title)
    #plt.subplot(gridSpec[plotPos])
    
# Fetching highest and lowest correlating pairs
def getExtremeCorrPairs(minCorrFactor, maxCorrFactor, corrTable):
    corrArray = np.array(corrTable)
    
    #minCorrFactor = 0.01
    #maxCorrFactor = 0.5
    
    lowCorrArray1 = np.argwhere(corrArray < minCorrFactor)
    highCorrArray1 = np.argwhere(corrArray > maxCorrFactor)
    
    def lex_sort(val1, val2, coeff, ixi, ixj):
        if ixi > ixj:
            ixi, ixj = ixj, ixi
        if val1 < val2:
            return (abs(coeff), val1,val2, ixi, ixj)
        else:
            return (abs(coeff), val2,val1, ixi, ixj)
    
    highCorrArray2 = [lex_sort(corrTable.columns[row[0]],corrTable.columns[row[1]], corrArray[row[0]][row[1]], row[0], row[1]) for row in highCorrArray1 if row[0] != row[1]]
    lowCorrArray2 = [lex_sort(corrTable.columns[row[0]],corrTable.columns[row[1]], corrArray[row[0]][row[1]], row[0], row[1]) for row in lowCorrArray1 if row[0] != row[1]]
    #print len(highCorrArray2)
    uniqueHighCorrList = dict.fromkeys(highCorrArray2).keys()
    uniqueHighCorrList.sort(reverse=True)
    
    uniqueLowCorrList = dict.fromkeys(lowCorrArray2).keys()
    uniqueLowCorrList = [tup for tup in uniqueLowCorrList if abs(tup[0]) < minCorrFactor]
    uniqueLowCorrList.sort()
    
    return (uniqueHighCorrList,uniqueLowCorrList,corrArray)

```

### Task 2 - Pair Wise correlation to identify the strongest and weakest correlating pairs

```python
# Main Module
datatable = loadTable()
corrTable = computeDataTableCorr(datatable, candidatesList)
displayCorrHeatGraph(corrTable, 'Correlation between different country attributes')
(uniqueHighCorrList,uniqueLowCorrList, corrArray) = getExtremeCorrPairs(0.01,0.5,corrTable)
#print('Corr Table : \n{}\n'.format(corrTable))
print('\n High Correlation List : \n')
dispFrame = DataFrame(uniqueHighCorrList, columns = ['CoEff', 'Attr1', 'Attr2', 'I1', 'I2'])
display(HTML(dispFrame[['CoEff', 'Attr1', 'Attr2']].to_html()))
print('\n Low Correlation List (Absoulute) : \n')
dispFrame = DataFrame(uniqueLowCorrList, columns = ['CoEff', 'Attr1', 'Attr2', 'I1', 'I2'])
display(HTML(dispFrame[['CoEff', 'Attr1', 'Attr2']].to_html()))
```

### Task 2 - Permuation test to determine p-value

```python
# Permutation test
testDataTable = datatable.copy(deep=True)
def shuffleDataTable(testDataTable):
    [np.random.shuffle(testDataTable[col]) for col in candidatesList]
   
shuffleDataTable(testDataTable)
shuffCorrTable = computeDataTableCorr(testDataTable, candidatesList)
#displayCorrHeatGraph(shuffCorrTable, 'Shuffle Test - Correlation between different country attributes')
#print('Shuff Corr Table : \n{}\n'.format(shuffCorrTable))
pValueArray = np.zeros(shuffCorrTable.shape)
shuffCorrTableSumArray = np.zeros(shuffCorrTable.shape)

def computePValues(testCount, pValueArray, testDataTable, candidatesList, corrArray, shuffCorrSumTable):
    init = False
    for i in range(1, testCount):
        shuffleDataTable(testDataTable)
        shuffCorrTable = computeDataTableCorr(testDataTable, candidatesList)
        shuffCorrArray = np.array(shuffCorrTable)
        #print('greater? {}'.format(shuffCorrArray>corrArray))
        if init == False:
            shuffCorrSumTable = shuffCorrTable
        else:
            shuffCorrSumTable += shuffCorrTable
        (m,n) = pValueArray.shape
        for mi in range(0,m):
            for ni in range(0,n):
                if math.pow(shuffCorrArray[mi,ni],2) > math.pow(corrArray[mi,ni],2):
                    pValueArray[mi,ni] += 1

    return (pValueArray/testCount,shuffCorrSumTable/testCount)

shuffCorrSumTable = pd.DataFrame()
(pValueArray,shuffCorrAvgTable) = computePValues(50, pValueArray, testDataTable, candidatesList, corrArray, shuffCorrSumTable)
displayCorrHeatGraph(shuffCorrAvgTable, 'Shuffle Test of 100 trials - Avg Correlation between different country attributes')
print("\n PValue Array for all correlations: \n")
dispFrame = DataFrame(pValueArray)
display(HTML(dispFrame.to_html()))
highCorrTupList = [(tup[1],tup[2],pValueArray[tup[3],tup[4]]) for tup in uniqueHighCorrList]
dispFrame = DataFrame(highCorrTupList, columns=['Attr1','Attr2','P-Value'])
print("\n PValue for highest correlating pairs: \n")
display(HTML(dispFrame.to_html()))
```

```python
# Linear Regression model - Predicting GDP per capita
import statsmodels.api as sm
from mpl_toolkits.mplot3d import Axes3D

TotalHealthExpPerCap = ((datatable['Health Exp/GDP'] / 100)*datatable['GDP (PPP) in US $'])/datatable['Population'] # corr 0.80
#LifeExp = datatable['Life Exp'] # corr 0.60
TotalMilitaryExpPerCap = ((datatable['Military Exp/GDP'] / 100)*datatable['GDP (PPP) in US $'])/datatable['Population']
GDPPerCap = datatable['GDP (PPP) in US $'] / datatable['Population']

y = GDPPerCap
#X = DataFrame([TotalHealthExpPerCap,LifeExp], index=['TotalHealthExpPerCap','LifeExp']).transpose()
X = DataFrame([TotalHealthExpPerCap,TotalMilitaryExpPerCap], index=['TotalHealthExpPerCap','TotalMilitaryExpPerCap']).transpose()
#X = X[['TotalHealthExpPerCap','LifeExp']]
X = X[['TotalHealthExpPerCap','TotalMilitaryExpPerCap']]
X = sm.add_constant(X)
est = sm.OLS(y, X).fit()

## Create the 3d plot -- skip reading this
# X grid for 3d plot
xx1, xx2 = np.meshgrid(np.linspace(X.TotalHealthExpPerCap.min(), X.TotalHealthExpPerCap.max(), 100), 
                       np.linspace(X.TotalMilitaryExpPerCap.min(), X.TotalMilitaryExpPerCap.max(), 100))
# plot the hyperplane by evaluating the parameters on the grid
Z = est.params[0] + est.params[1] * xx1 + est.params[2] * xx2

# create matplotlib 3d axes
fig = plt.figure(figsize=(12, 8))
ax = Axes3D(fig, azim=-115, elev=15)

# plot hyperplane
surf = ax.plot_surface(xx1, xx2, Z, cmap=plt.cm.RdBu_r, alpha=0.6, linewidth=0)

# plot data points - points over the HP are white, points below are black
resid = y - est.predict(X)
ax.scatter(X[resid >= 0].TotalHealthExpPerCap, X[resid >= 0].TotalMilitaryExpPerCap, y[resid >= 0], color='black', alpha=1.0, facecolor='white')
ax.scatter(X[resid < 0].TotalHealthExpPerCap, X[resid < 0].TotalMilitaryExpPerCap, y[resid < 0], color='black', alpha=1.0)

# set axis labels
ax.set_xlabel('TotalHealthExpPerCap')
ax.set_ylabel('TotalMilitaryExpPerCap')
ax.set_zlabel('GDPPerCap')

Summary = est.summary()
Z = est.params[0] + (est.params[1] * TotalHealthExpPerCap) + (est.params[2] * TotalMilitaryExpPerCap)
ZIncOvery = (Z-y)/y*100
ZDiffOvery = Z-y
RegCompTable = pd.concat([datatable['Country Name'], y, Z, ZIncOvery, ZDiffOvery], axis=1, keys=['Country','ActVal','EstVal','PercntInc','Diff'])
RegPercCompTable = RegCompTable.sort(columns='PercntInc', ascending=False)
RegDiffCompTable = RegCompTable.sort(columns='Diff', ascending=False)
print("\n Countries Most Above Forecast in terms of % Inc: \n")
display(HTML(RegPercCompTable.head(5).to_html()))
print("\n Countries Most Below Forecast in terms of % Inc: \n")
display(HTML(RegPercCompTable.tail(5).to_html()))
print("\n Countries Most Above Forecast in terms of Diff: \n")
display(HTML(RegDiffCompTable.head(5).to_html()))
print("\n Countries Most Below Forecast in terms of Diff Inc: \n")
display(HTML(RegDiffCompTable.tail(5).to_html()))
```

```python
# World Heat Map to analyse the Life Expectancy
import pylab as p
from mpl_toolkits.basemap import Basemap
from matplotlib.colors import rgb2hex
import random


mapMigFrame = vizFrame[['Country Name','Life Exp']].copy(deep=True)
mapMigFrame = mapMigFrame.dropna()

plt.figure(figsize=(18,9))
m = Basemap(llcrnrlon=-180,llcrnrlat=-90,urcrnrlon=180,urcrnrlat=90,
           projection='mill')
#m = Basemap(projection='robin', resolution = 'h', area_thresh = 1000.0,
 #             lat_0=0, lon_0=-130)
# draw country boundaries.
# data from http://www.cipotato.org/DIVA/data/MoreData.htm
shp_info = m.readshapefile('C:\Stuffs\StonyBrook\Courses\DataScience\WorldAdm\world_adm0','countries',drawbounds=True)
# choose a color for each state based on population density.
colors={}
countrynames=[]
cmap = p.cm.jet # use 'jet' colormap
vmin = min(mapMigFrame['Life Exp']); vmax = max(mapMigFrame['Life Exp']) # set range of Welfare

#shapedict format
#{'RINGNUM': 9, 'REGION': 'North America', 'NAME': 'United States', 'SHAPENUM': 197, 'GMI_CNTRY': 'USA'}
#{'RINGNUM': 10, 'REGION': 'North America', 'NAME': 'United States', 'SHAPENUM': 197, 'GMI_CNTRY': 'USA'}

for shapedict in m.countries_info:
    mapcountryname = shapedict['NAME']
    
    CMigScore = mapMigFrame.loc[(mapMigFrame['Country Name'] == mapcountryname)]
    
    try:
        MigScore = CMigScore.iloc[0]['Life Exp']
        
        # calling colormap with value between 0 and 1 returns
        # rgba value.  Invert color range (hot colors are high
        # population), take sqrt root to spread out colors more.
        colors[mapcountryname] = cmap(1.-p.sqrt((MigScore-vmin)/(vmax-vmin)))[:3]
    except:
        colors[mapcountryname] = -1
    countrynames.append(mapcountryname)
#print countrynames
# cycle through country names, color each one.
for nshape,seg in enumerate(m.countries):
    xx,yy = zip(*seg)
    if colors[countrynames[nshape]] == -1:
        p.fill(xx,yy,'white',edgecolor='white')
    else:
        color = rgb2hex(colors[countrynames[nshape]])
        p.fill(xx,yy,color,edgecolor=color)
# draw meridians and parallels.
m.drawparallels(p.arange(-90,91,30),labels=[1,0,0,0])
m.drawmeridians(p.arange(-180,181,60),labels=[0,0,0,1])
p.title('Life Expectency Analysis across countries')
p.show()
```

##### Although most of the world countries have higher literacy rate, it is seen that the African countries lag far behind. Hence it can be inferred that there is lack of awareness for importance of education in the African countries. This is one of the major factor affecting the development of the African countries.
